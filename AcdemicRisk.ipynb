```python
# make a copy Data first
df = data.copy()
```

```python
# Check Missing Value
print(df.isnull().sum())
```

```python
# fill numeric missing values with median
df["FamChallenge"] = df["FamChallenge"].fillna(df["FamChallenge"].median())

# fill categorical missing values with mode
df["IF_AIUse"] = df["IF_AIUse"].fillna(df["IF_AIUse"].mode()[0])

# check again
print(df.isnull().sum())
```

## **2.3 FEATURE SELECT**



```python
# Drop Feature
df.drop('ID', axis=1, inplace=True)
```

```python
#check Data Frame and Feature
df.head()
```

```python
#Check Missing Value
df.isnull().sum()
```

```python
for col in ["Midterm", "Final", "Attendance", "HWComplete"]:
    print(col, df[col].min(), df[col].max())
```

```python
# Check min Max of ["Midterm", "Final", "Attendance", "HWComplete"]
cols = ["Midterm", "Final", "Attendance", "HWComplete"]
print(df[cols].describe())
for col in cols:
    print(col, "min:", df[col].min(), "max:", df[col].max())
```

```python
# 1. Distribution of source variables
print(df[["Midterm", "Final", "Attendance", "HWComplete"]].describe())
```

```python
# Create warning flags
df["MidtermWarning"] = (df["Midterm"] < 50).astype(int)
df["FinalWarning"] = (df["Final"] < 50).astype(int)
df["AttendanceWarning"] = (df["Attendance"] < 75).astype(int)
df["HWWarning"] = (df["HWComplete"] < 60).astype(int)

# Create risk score
df["RiskScore"] = (
    df["MidtermWarning"]
    + df["FinalWarning"]
    + df["AttendanceWarning"]
    + df["HWWarning"]
)

# Create final target variable
df["AcademicRisk"] = (df["RiskScore"] >= 2).astype(int)

# Check result
print(df[[
    "Midterm", "Final", "Attendance", "HWComplete",
    "MidtermWarning", "FinalWarning", "AttendanceWarning", "HWWarning",
    "RiskScore", "AcademicRisk"
]].head())

# Check class distribution
print(df["AcademicRisk"].value_counts())
print(df["AcademicRisk"].value_counts(normalize=True) * 100)
```
